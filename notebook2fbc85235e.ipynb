```python
cars = pd.read_csv('/kaggle/input/cardekho-used-car-data/cardekho_dataset.csv')
```

```python
cars.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
cars = cars.drop(columns = ['brand','model'], axis = 1)
```

```python
ccount = cars['car_name'].value_counts()
```

```python
ccount = ccount[ccount > 50]
```

```python
cars['car_name'] = cars['car_name'].apply(lambda s : None if s not in ccount else s)
```

```python
ccount = cars['vehicle_age'].value_counts()
ccount = ccount[ccount > 15]
```

```python
cars['vehicle_age'] = cars['vehicle_age'].apply(lambda s : None if s not in ccount else s)
```

```python
cars['vehicle_age'].value_counts()
```

```python
ccount = cars['km_driven'].value_counts()
ccount = ccount[ccount > 10]
```

```python
cars['km_driven'] = cars['km_driven'].apply(lambda s : None if s not in ccount else s)
```

```python
cars
```

```python
cars['seller_type'] = cars['seller_type'].apply(lambda s : 0 if s == 'Dealer' else 1)
```

```python
cars['transmission_type'] = cars['transmission_type'].apply(lambda s : 0 if s == 'Manual' else 1)
```

```python
ccount = cars['mileage'].value_counts()
ccount = ccount[ccount < 10]
```

```python
cars['mileage'] = cars['mileage'].apply(lambda x : None if x in ccount else x)
```

```python
ccount = cars['engine'].value_counts()
ccount = ccount[ccount < 10]
```

```python
cars['engine'] = cars['engine'].apply(lambda x : None if x in ccount else x)
```

```python
ccount = cars['max_power'].value_counts()
ccount = ccount[ccount < 10]
```

```python
cars['max_power'] = cars['max_power'].apply(lambda x : None if x in ccount else x)
```

```python
ccount = cars['seats'].value_counts()
ccount = ccount[ccount < 10]
```

```python
cars['seats'] = cars['seats'].apply(lambda x : None if x in ccount else x)
```

```python
cars = cars.dropna()
cars = cars.drop_duplicates()
```

```python
cars
```

```python
cars['car_name'].value_counts()
```

```python
features = cars.drop(columns=['selling_price'])
targets = cars['selling_price']
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score
```

```python
features_train, features_test, targets_train, targets_test = train_test_split(features, targets, test_size = 1/3, random_state = 0)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse=False),['car_name','fuel_type']), remainder='passthrough')
scaler = StandardScaler()
lr = Ridge()
pipe = make_pipeline(column_trans, scaler, lr)
pipe.fit(features_train, targets_train)
```

```python
y_pred = pipe.predict(features_test)
print('R2 score = ',r2_score(targets_test,y_pred))
```
